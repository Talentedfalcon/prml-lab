````python
heart_disease=pd.read_csv('./heart-disease.csv')
heart_disease.head()
````

````python
heart_disease["age"].plot.hist(bins=50)
````

````python
heart_disease.plot.hist(figsize=(5,20),subplots=True)
````

````python
over_50=heart_disease[heart_disease["age"]>50]
over_50
````

````python
over_50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,6))
````

````python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax)
ax.set_xlim([45,100])
````

````python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(over_50["age"],over_50["chol"],c=over_50["target"])

ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

ax.legend(*scatter.legend_elements(),title="Target")
````

````python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(over_50["age"],over_50["chol"],c=over_50["target"])
ax.set(title="Heart Disease and Cholestrol Level",
       xlabel="Age",
       ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),
          title="Target")
ax.axhline(over_50["chol"].mean(),
           linestyle="--")
````

````python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,8))

scatter=ax0.scatter(over_50["age"],over_50["chol"],c=over_50["target"])

ax0.set(title="Heart Disease and Cholestrol Level",ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),color='b',linestyle='--',label="Average")

scatter=ax1.scatter(over_50["age"],over_50["thalach"],c=over_50["target"])

ax1.set(title="Heart Disease and Cholestrol Level",xlabel="Age",ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(),title="Target")

ax1.axhline(y=over_50["thalach"].mean(),color='b',linestyle='--',label="Average")

fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight='bold')
````

````python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,8))

scatter=ax0.scatter(over_50["age"],over_50["chol"],c=over_50["target"],cmap="winter")

ax0.set(title="Heart Disease and Cholestrol Level",ylabel="Cholestrol",xlim=[50,80])
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),color='b',linestyle='--',label="Average")

scatter=ax1.scatter(over_50["age"],over_50["thalach"],c=over_50["target"],cmap="winter")

ax1.set(title="Heart Disease and Cholestrol Level",xlabel="Age",ylabel="Max Heart Rate",ylim=[60,220])
ax1.legend(*scatter.legend_elements(),title="Target")

ax1.axhline(y=over_50["thalach"].mean(),color='b',linestyle='--',label="Average")

fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight='bold')
````
