```python
import numpy as np
```

```python
def MatrixAddition(mat1,mat2):
    return mat1+mat2
def MatrixSubtraction(mat1,mat2):
    return mat1-mat2
def ScalarMatrixMultiplication(scalar,mat):
    return scalar*mat
def ElementwiseMatrixMultiplication(mat1,mat2):
    return mat1*mat2
def MatrixMultiplication(mat1,mat2):
    return np.dot(mat1,mat2)
def MatrixTranspose(mat):
    return np.transpose(mat)
def TraceofMatrix(mat):
    return np.trace(mat)
def SolveLinearEquation(A,B):
    A_inv=Inverse(A)
    return MatrixMultiplication(A_inv,B)
def Determinant(mat):
    return np.linalg.det(mat)
def Inverse(mat):
    return np.linalg.inv(mat)
def Eigen(mat):
    return np.linalg.eig(mat)
```

```python
def mat_input():
    num_rows,num_cols=[int(i) for i in input("rxc").split('x')]
    print(num_rows,num_cols)
    arr=[int(i) for i in input("Elements").split(" ")]
    if(len(arr)!=num_cols*num_rows):
        raise Exception("Number of elements is invalid!")
    return np.array(arr).reshape(num_rows,num_cols)
```
