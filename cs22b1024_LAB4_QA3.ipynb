```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import cv2
```

```python
def calculate_histogram(image_path, bins=256):
    image = Image.open(image_path)
    image_array = np.array(image)
    histogram = cv2.calcHist([image_array], [0], None, [bins], [0, 256])
    cv2.normalize(histogram, histogram)
    return histogram
```

```python
fig,ax=plt.subplots(3)
query_hist=calculate_histogram('./pictures/queryimage.png')
ax[0].hist(query_hist)
ax[0].set_title("Query Image")
ref1_hist=calculate_histogram('./pictures/refimage1.png')
ax[1].hist(ref1_hist)
ax[1].set_title("Ref Image 1")
ref2_hist=calculate_histogram('./pictures/refimage2.png')
ax[2].hist(ref2_hist)
ax[2].set_title("Ref Image 2")

fig.tight_layout()
plt.show()
```

```python
print(f"The Battacharya Distance between queryimage and refimage1: {cv2.compareHist(query_hist,ref1_hist,cv2.HISTCMP_BHATTACHARYYA)}")
print(f"The Battacharya Distance between queryimage and refimage2: {cv2.compareHist(query_hist,ref2_hist,cv2.HISTCMP_BHATTACHARYYA)}")
```
