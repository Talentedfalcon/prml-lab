```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h1 = np.array([ 0.24, 0.2, 0.16, 0.12, 0.08, 0.04, 0.12, 0.04])
h2 = np.array([ 0.22, 0.23, 0.16, 0.13, 0.11, 0.08, 0.05, 0.02])

fig,ax=plt.subplots(2)
ax[0].hist(h1)
ax[0].set_title("Hist1");
ax[1].hist(h2)
ax[1].set_title("Hist2");
fig.tight_layout()
```

```python
def KL_distance(hist1,hist2):
    return np.sum(hist1*np.log2(hist1/hist2))

def Battacharya_distance(hist1,hist2):
    return -np.log(np.sum(np.sqrt(hist1*hist2)))

def Cosine_distance(a,b):
    return 1-(np.sum(a*b)/(np.sqrt(np.sum(a**2))*np.sqrt(np.sum(b**2))))
```

```python
print(f"KL Distance = {KL_distance(h1,h2)}")
print(f"Battacharya Distance = {Battacharya_distance(h1,h2)}")
print(f"Cosine Distance = {Cosine_distance(h1,h2)}")
```
