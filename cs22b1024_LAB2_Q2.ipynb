```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
x
```

```python
fig,ax=plt.subplots(2,2) #Create the 4 subplots
fig.set_figheight(10)
fig.set_figwidth(12)
plt.show()
```

```python
ax[0][0].cla()
ax[0][0].plot(x,x,label="line1",linewidth=0.5) #y=x
ax[0][0].plot(x,x**2,label="line2",linewidth=1) #y=x^2
ax[0][0].plot(x,x**3,label="line3",linewidth=2) #y=x^3
ax[0][0].legend()
ax[0][0].set_title("Basic Line Plots")

display(fig)
```

```python
ax[0][1].cla()
ax[0][1].plot(x,np.sin(x)+4,label="solid",linestyle="solid",color="orange")  #Adding a small increment to the sin(x) to keep them from overlapping
ax[0][1].plot(x,np.sin(x)+2,label="dashed",linestyle="dashed",color="orange")
ax[0][1].plot(x,np.sin(x),label="dashdot",linestyle="dashdot",color="orange")
ax[0][1].plot(x,np.sin(x)-2,label="dotted",linestyle="dotted",color="orange")

ax[0][1].legend()
ax[0][1].set_title("Line Styles")

display(fig)
```

```python
ax[1][0].cla()
ax[1][0].plot(x,np.cos(x)+4,label="+ marker",marker="+",markersize=5,color="pink")  #Adding a small increment to the cos(x) to keep them from overlapping
ax[1][0].plot(x,np.cos(x)+2,label="o marker",marker="o",markersize=8,color="pink")
ax[1][0].plot(x,np.cos(x),label="s marker",marker="s",markersize=11,color="pink")
ax[1][0].plot(x,np.cos(x)-2,label="^ marker",marker="^",markersize=14,color="pink")

ax[1][0].legend()
ax[1][0].set_title("Markers")

display(fig)
```

```python
ax[1][1].cla()
line1,=ax[1][1].plot(x,np.tan(x),marker="o",markerfacecolor="yellow",markeredgecolor="red")
line1.set_dashes([2,4,8,4]) # 2pt of line, 4pts of break, 8pts of line and 4pts of break

ax[1][1].set_title("Custom Line and Marker Styles")

display(fig)
```

```python
for i in ax:
    for j in i:
        j.set_xlabel("x")
        j.grid()

ax[0][0].set_ylabel("x")
ax[0][1].set_ylabel("sin(x)")
ax[1][0].set_ylabel("cos(x)")
ax[1][1].set_ylabel("tan(x)")

display(fig)
```

```python
fig.suptitle("Matplotlib Line Plot Customization")
display(fig)
```
