Column Operations

```python
cols=heart[['Age','Sex']]
cols
```

```python
def AgeBased(x):
    if(x<50):
        return 'Young'
    elif(x<=70):
        return 'Middle-aged'
    else:
        return 'Senior'

heart['AgeGroup']=heart['Age'].apply(AgeBased)
heart
```

Conditional Statements

Handling Missing Values

```python
heart.isnull().sum().sum()
```

```python
for column in heart.columns:
    if(pd.api.types.is_numeric_dtype(heart[column])):
        heart[column]=heart[column].fillna(heart[column].mean())
    elif(pd.api.types.is_object_dtype(heart[column])):
        heart[column]=heart[column].fillna(heart[column].mode()[0])

heart
```

Concatenating Data

```python
heart_male=heart[(heart['Sex']=='M')].copy()
heart_female=heart[(heart['Sex']=='F')].copy()

display(heart_male)
display(heart_female)
```

```python
heart_male['Group']='Male'
heart_female['Group']='Female'

display(heart_male)
display(heart_female)
```

```python
heart=pd.concat([heart_male,heart_female])
heart
```

Applying Functions

```python
def BMI(row):
    return (row['Weight (kg)']/(row['Height (m)']**2))

heart['Weight (kg)']=np.random.randint(30,100,heart.shape[0])
heart['Height (m)']=np.random.randint(1.2,3.0,heart.shape[0])

heart['BMI']=heart.apply(BMI,axis=1)

heart
```

```python
heart['AgeTime100']=heart['Age'].apply(lambda x: x*100)
heart
```
